```python
import pandas as pd
all_data_M_2022 = pd.read_excel("all_data_M_2022.xlsx", dtype = 'str')
CIP_SOC_Crosswalk = pd.read_excel("CIP2020_SOC2018_Crosswalk.xlsx", sheet_name = "CIP-SOC", dtype = 'str')
CIP_SOC_Crosswalk['CIPFamily'] = CIP_SOC_Crosswalk['CIP2020Code'].str[:2]
CIP_SOC_Crosswalk['CIP2020Code'] = CIP_SOC_Crosswalk['CIP2020Code'].str.replace('.','-')
all_data_M_2022['SOC2018Code'] = all_data_M_2022['OCC_CODE']
del all_data_M_2022['OCC_CODE']
pseo = pd.read_csv("merged_df.csv")
pseo = pseo.drop(pseo.columns[0],axis = 1)
```

```python

institution_all = pseo['institution'].unique().tolist()
degree_level_all = pseo['degree_level'].unique().tolist()
cip_name_all = pseo['cip_name'].unique().tolist()
state_all = pseo['institution_state'].unique().tolist()
industry_all = pseo['industry_name'].unique().tolist()
import tkinter as tk
from tkinter import Listbox, Scrollbar, StringVar, Button

# 创建Tkinter窗口
root = tk.Tk()
root.title("选择选项")

# 创建五个空的列表，用于存储用户的选择
selected_institution = []#2
selected_degree_level = []
selected_cip_name = []#3
selected_state = []#1
selected_industry = []#4

submitted_institution = []
submitted_degree_level = []
submitted_cip_name = []
submitted_state = []
submitted_industry = []

# 创建Tkinter变量来跟踪用户的选择
var_institution = StringVar()
var_degree_level = StringVar()
var_cip_name = StringVar()
var_state = StringVar()
var_industry = StringVar()

# 创建五个Listbox小部件，横向排布
listbox_institution = Listbox(root, listvariable=var_institution, selectmode=tk.MULTIPLE, width=30)
listbox_degree_level = Listbox(root, listvariable=var_degree_level, selectmode=tk.MULTIPLE, width=30)
listbox_cip_name = Listbox(root, listvariable=var_cip_name, selectmode=tk.MULTIPLE, width=30)
listbox_state = Listbox(root, listvariable=var_state, selectmode=tk.MULTIPLE, width=30)
listbox_industry = Listbox(root, listvariable=var_industry, selectmode=tk.MULTIPLE, width=30)

# 放置Listbox小部件
listbox_institution.grid(row=0, column=6, padx=20, pady=10)
listbox_degree_level.grid(row=0, column=2, padx=10, pady=10)
listbox_cip_name.grid(row=0, column=4, padx=10, pady=10)
listbox_state.grid(row=0, column=0, padx=5, pady=10)
listbox_industry.grid(row=0, column=8, padx=10, pady=10)

# 添加滚动条
scrollbar_institution = Scrollbar(root, command=listbox_institution.yview)
scrollbar_degree_level = Scrollbar(root, command=listbox_degree_level.yview)
scrollbar_cip_name = Scrollbar(root, command=listbox_cip_name.yview)
scrollbar_state = Scrollbar(root, command=listbox_state.yview)
scrollbar_industry = Scrollbar(root, command=listbox_industry.yview)

scrollbar_institution.grid(row=0, column=1, sticky='ns')
scrollbar_degree_level.grid(row=0, column=3, sticky='ns')
scrollbar_cip_name.grid(row=0, column=5, sticky='ns')
scrollbar_state.grid(row=0, column=7, sticky='ns')
scrollbar_industry.grid(row=0, column=9, sticky='ns')

listbox_institution.config(yscrollcommand = scrollbar_institution.set)
listbox_degree_level.config(yscrollcommand = scrollbar_degree_level.set)
listbox_cip_name.config(yscrollcommand = scrollbar_cip_name.set)
listbox_state.config(yscrollcommand = scrollbar_state.set)
listbox_industry.config(yscrollcommand = scrollbar_industry.set)

# 设置可选项

var_institution.set(tuple(institution_all))
var_degree_level.set(tuple(degree_level_all))
var_cip_name.set(tuple(cip_name_all))
var_state.set(tuple(state_all))
var_industry.set(tuple(industry_all))

    

# 定义选择回调函数
def select_institution(event):
    for index in listbox_institution.curselection():
        selected_institution.append(institution_all[index])
    for i in selected_institution:
        if i not in submitted_institution:
            submitted_institution.append(i)
    print("Submitted Institution: ", submitted_institution)

def select_degree_level(event):
    for index in listbox_degree_level.curselection():
        selected_degree_level.append(degree_level_all[index])
    for i in selected_degree_level:
        if i not in submitted_degree_level:
            submitted_degree_level.append(i)
    print("Submitted Degree Level: ", submitted_degree_level)

def select_cip_name(event):
    for index in listbox_cip_name.curselection():
        selected_cip_name.append(cip_name_all[index])
    for i in selected_cip_name:
        if i not in submitted_cip_name:
            submitted_cip_name.append(i)
    print("Submitted CIP Name: ", submitted_cip_name)
    
def select_state(event):
    for index in listbox_state.curselection():
        selected_state.append(state_all[index])
    for i in selected_state:
        if i not in submitted_state:
            submitted_state.append(i)
    print("Submitted State: ", submitted_state)
    print("Submitted CIP Name: ", submitted_cip_name)
    
def select_industry(event):
    for index in listbox_industry.curselection():
        selected_industry.append(industry_all[index])
    for i in selected_industry:
        if i not in submitted_industry:
            submitted_industry.append(i)
    print("Submitted Industyr: ", submitted_industry)

def update_institution():
    global var_institution
    filtered_df = pseo
    if submitted_state: 
        submitted_states_set = set(submitted_state)
        filtered_df = filtered_df[filtered_df['institution_state'].isin(submitted_states_set)]   
    if submitted_degree_level:
        submitted_degree_level_set = set(submitted_degree_level)
        filtered_df = filtered_df[filtered_df['degree_level'].isin(submitted_degree_level_set)]
    if submitted_cip_name:
        submitted_cip_name_set = set(submitted_cip_name)
        filtered_df = filtered_df[filtered_df['cip_name'].isin(submitted_cip_name_set)]
    if submitted_industry:
        submitted_industry_set = set(submitted_industry)
        filtered_df = filtered_df[filtered_df['industry'].isin(submitted_industry_set)]
    filtered_institutions = filtered_df['institution'].unique().tolist()
    var_institution.set(tuple(filtered_institutions))
    print(var_institution.get())
    


    
def submit_state(event):
    update_institution()

#def submit_institution(event):

def submit_degree_level(event):
    update_institution()

def submit_cip_name(event):
    update_institution()

def submit_industry(event):
    update_institution()
    
# 绑定选择回调函数
listbox_institution.bind('<<ListboxSelect>>', select_institution)
listbox_degree_level.bind('<<ListboxSelect>>', select_degree_level)
listbox_cip_name.bind('<<ListboxSelect>>', select_cip_name)
listbox_state.bind('<<ListboxSelect>>', select_state)
listbox_industry.bind('<<ListboxSelect>>', select_industry)



# 添加提交按钮
submit_button_institution = Button(root, text="提交 Institution 选择")
submit_button_degree_level = Button(root, text="提交 Degree Level 选择")
submit_button_cip_name = Button(root, text="提交 CIP Name 选择")
submit_button_state = Button(root, text="提交 State 选择")
submit_button_industry = Button(root, text="提交 Industry 选择")

submit_button_institution.grid(row=1, column=6, padx=10, pady=10)
submit_button_degree_level.grid(row=1, column=2, padx=10, pady=10)
submit_button_cip_name.grid(row=1, column=4, padx=10, pady=10)
submit_button_state.grid(row=1, column=0, padx=10, pady=10)
submit_button_industry.grid(row=1, column=8, padx=10, pady=10)

submit_button_state.bind('<Button-1>', submit_state)
submit_button_institution.bind('<Button-1>', submit_institution)
submit_button_degree_level.bind('<Button-1>', submit_degree_level)
submit_button_cip_name.bind('<Button-1>', submit_cip_name)
submit_button_industry.bind('<Button-1>', submit_industry)

# 启动Tkinter主循环
root.mainloop()
```
